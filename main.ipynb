```python
from openai import OpenAI 
import docx
import os
```

```python
OpenAI.api_key = os.getenv("OPENAI_API_KEY")
```

```python
client = OpenAI(
    api_key = OpenAI.api_key
)
```

```python
def generate_questions(prompt, model="gpt-4o"):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "You are a helpful assistant that generates exam questions."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )
    return response.choices[0].message.content

# 문제 출제를 위한 프롬프트 작성
prompt = """
아래 정보에 기반하여 서로 다른 문제 10개를 출제해줘.

## 문제의 형식
1. 문제는 발문과 선지로 이루어져 있다.
2. 선지는 반드시 4개이고 중복 정답은 없다.
3. 모든 문제는 '모든 답이 맞다.' 혹은 '모든 답이 틀리다.', '정답이 없다.'라는 선지를 가질 수 없다.
4. 각 선지에 발문의 내용과 크게 상충되거나, 극단적으로 보여지는 내용은 사용하지 않는다.
5. 위 네가지 형식을 반드시 준수하되, 숫자가 적은 형식이 우위를 가진다.


## 문제 예시
1. 헤어세팅에 있어서 웨이브의 형상에 따라 분류하는 것으로서 크레스트가 너무 약하게 되어 리지가 눈에 잘 띄지 않은 웨이브는?

1) 버티컬 웨이브    2) 새도우 웨이브    3) 내로우 웨이브    4) 와이드 웨이브
정답: 새도우 웨이브


## 출제 범위
피부 미용이란

피부 미용은 피부의 건강, 청결, 외모를 개선하고 유지하는 것을 목적으로 하는 활동과 관행을 포함 한다. 이 분야는 일상적인 피부 관리 루틴, 전문적인 피부 치료, 그리고 피부 건강에 영향을 미치는 생활 습관에 이르기까지 다양한 측면을 포괄한다. 피부 미용의 주요 목표는 피부의 건강을 증진시 키고, 피부 문제를 예방하거나 치료하며, 전반적인 외모를 개선하는 것이다.
피부 미용은 단순히 외모를 가꾸는 것 이상의 의미를 지니며, 개인의 건강, 웰빙, 자존감에 중요한 역할을 한다. 이는 건강한 피부를 유지하기 위한 실질적인 방법과 일상의 웰빙을 향상시키는 정신 적인 측면 모두를 포함하는 광범위한 개념이다.


피부 미용의 중요성

피부 건강 유지
피부는 인체에서 가장 큰 기관으로서, 외부 환경으로부터 보호하는 중요한 역할을 한다. 적절 한 피부 관리는 피부의 기능을 유지하고 각종 피부 질환을 예방하는 데 도움을 준다.

자신감 향상
건강하고 매력적인 피부는 개인의 자신감과 자존감에 긍정적인 영향을 미친다. 피부 미용은 외 모에 대한 만족감을 높여, 개인의 사회적 상호작용과 전반적인 삶의 질을 향상시킬 수 있다.

노화 방지
적절한 피부 관리는 피부 노화를 늦추는 데 도움이 된다. 선크림 사용, 보습, 적절한 영양 섭취 등은 주름과 기타 노화 징후를 최소화하는 데 기여하였다.

스트레스 해소
피부 관리 루틴은 일종의 자기 관리 활동으로, 스트레스 해소에도 도움이 된다. 이러한 루틴은 명상적인 효과를 제공하여 마음의 안정을 찾는 데에도 기여할 수 있다.

건강한 생활 습관의 촉진
피부 미용에 대한 관심은 종종 건강한 생활 습관으로 이어진다. 피부 건강에 좋은 식습관, 충분 한 수면, 적절한 수분 섭취 등은 전반적인 건강에도 긍정적인 영향을 미친다.




피부 미용의 용어와 영역

코스메틱 (Cosmetic / 영국, Kosmetik / 독일)
유래 : 코스메틱은 그리스어 ‘Kosmos, kosmein’에서 유래되었다. 이 단어들은 ‘우주, 장식, 조 화’와 같은 개념을 나타낸다.
의미 : 코스메틱은 아름다움과 건강을 추구하는 신체 관리를 조화롭고 질서 있는 방식으로 다루는 것을 의미한다. 이는 메이크업, 스킨케어, 헤어케어 제품 등을 포함하여, 외적인 아름다움 을 증진시키는데 사용되는 다양한 제품을 지칭한다.

에스테틱 (Esthetic / 미국·한국, Aesthetic / 미국)
유래 : 에스테틱은 프랑스어 ‘Esthetique’에서 유래되었다. 이 단어는 ‘미학의, 미의, 심미적인’을 뜻한다.
의미 : 이 용어는 얼굴과 신체 피부의 기능을 정상화하고 유지하는 데 초점을 맞춘 전신 미용술을 의미한다. 부의 건강과 아름다움을 동시에 추구하며, 피부 관리, 마사지, 바디 트리트먼트 등의 서비스를 포함한다.

에스테 띠끄 (Esthetique)
의미 : 에스테 띠끄는 피부 관리와 화장품을 구별하기 위해 사용되는 용어로, 에스테틱과 유사한 의미를 가진다. 이는 헤어스타일링을 제외한 모든 미용 관리를 포함한다. 따라서, 얼굴과 몸 의 피부 관리,피부 건강, 미용 트리트먼트 등을 포함한다.

에스테 티션 (Esthetician)
의미 : 에스테 티션은 피부 미용을 전문적으로 수행하는 사람을 지칭한다. 이들은 피부미용사 또 는 에스테티션이라고 불리며, 피부 관리, 피부 상태 진단, 트리트먼트, 미용 조언 등을 제공 한다.
역할 : 에스테티션은 개인의 피부 타입과 상태에 맞는 피부 관리 방법과 제품을 추천하고, 전문적 인 피부관리 서비스를 제공한다. 이들은 또한 고객에게 피부 건강과 관련된 교육과 상담을 제공하기도 한다

안면관리
피부유형을 분석하고 유형별 특징에 맞게 관리
미백, 보습, 재생, 여드름 등 다양한 관리 방법을 안면 부위에 적용

전신관리
안면 부위를 제외한 전신을 관리 영역으로 구분함
관리 목적에 따라 수기요법, 림프드레나쥐 등을 신체 각 부위에 적용

특수관리
피부에 유효성이 있는 화장품, 미용기기를 이용하여 관리

메디컬 관리
의학적 치료와 피부관리가 결합된 것으로 의사가 관리 대상자를 고려하여 피부 를 진단하고 의료기기를 이용하여 치료, 관리



피부 미용의 위생관리

1. 위생관리 자세

1) 전문적인 이미지 유지
항상 청결하고 단정한 복장을 유지하고 언어 사용과 표정 관리를 통해 전문성을 나타난다.
개인 위생을 철저히 지키며, 구취나 체취가 나지 않도록 주의한다.

2) 손 위생의 중요성
관리 전, 후에 수시로 손을 씻고 소독하여 청결을 유지한다.
손톱은 짧고 깔끔하게 정리하며, 색깔 있는 네일 폴리시 사용을 자제한다.

3) 작업 중 위생 관리
관리 중 전화 수신이나 타인 물건 접촉(자신의 머리카락 포함) 후에는 반드시 소독하여 위생 을 유지한다.

4) 적절한 복장과 신발 착용
편안하고 소음이 없는 흰색 신발을 착용하여, 신발 소리가 나지 않도록 주의한다.

5) 머리카락과 화장 관리
긴 머리는 단정하게 묶어 올리고, 자연스러운 화장을 한다.

6) 장신구 착용 자제
관리 중에는 목걸이 , 반지, 팔찌와 같은 장신구 착용을 자제한다.


2. 작업환경

1) 청결과 위생
작업환경은 항상 깨끗하고 위생적으로 유지되어야 합니다. 모든 표면, 기구, 장비는 정기적으 로 청소하고 소독해야 한다
공간은 먼지와 오염물이 없어야 하며, 사용하는 모든 도구는 깨끗하고 상태가 좋아야 한다.

2) 조명과 온도
충분한 조명은 중요하지만, 너무 밝거나 강한 조명을 피해야 한다. 조명은 부드럽고 편안한 분 위기를 조성할 수 있다.

3) 소음 관리
작업 환경은 조용하고 평온해야 한다. 외부 소음이 최소화되어야 하며, 필요한 경우 음악을 사 용해 분위기를 조정할 수 있다.
대화나 기타 소리는 저음으로 유지하고, 소음이 발생하는 장비는 신중히 배치해야 한다.

4) 공간 배치와 편의성
작업 공간은 넓고 편안해야 하며, 충분한 이동 공간을 확보해야 한다.
모든 필요한 기구와 제품은 사용하기 편리한 위치에 배치되어야 한다.
고객과 직원 모두의 편의를 위해 잘 설계된 공간과 탈의실, 화장실 등의 부대 시설이 필요

5) 안전
모든 전기 장비와 기구는 안전하게 사용되고 관리되어야 한다.
비상 상황에 대비한 안전 장비가 쉽게 접근 가능한 곳에 배치되어 있어야 한다.

6) 개인 공간의 존중
고객의 프라이버시와 편안함을 최우선으로 고려한다. 치료 중 개인적인 대화는 자제하고, 필요한 경우 개인 공간을 제공한다.


3. 작업환경/도구 위생관리

1) 정기적인 청소 및 소독
모든 작업 공간, 치료대, 장비 및 도구는 매일 정기적으로 청소하고 소독해야 한다.
바닥, 표면, 손잡이, 스위치 등 자주 접촉되는 곳은 특히 깨끗이 유지한다.

2) 도구 및 장비 관리
모든 사용 도구는 사용 후 즉시 청소하고, 필요한 경우 소독한다.
일회용 도구는 사용 후 적절하게 폐기한다.
재사용 가능한 도구는 올바르게 소독하고 건조시킨 후 보관한다.

3) 개인 보호 장비(PPE) 사용
필요에 따라 마스크, 장갑, 보호 안경 등의 개인보호 장비를 착용한다.
PPE는 올바르게 착용하고, 사용 후 적절하게 폐기하거나 청소한다.

4) 손 위생
관리자와 직원 모두는 작업 전후로 꼭 손을 씻고, 필요한 경우 손 소독제를 사용한다.
손톱은 짧게 유지하고, 손톱 밑과 손가락 사이의 청결도 주의 깊게 관리한다.

5) 공기질 관리
적절한 환기를 통해 실내 공기의 질을 유지한다.
공기 정화기 또는 여과 시스템의 사용을 고려한다.

6) 쓰레기 관리
쓰레기통은 뚜껑이 닫히는 유형을 사용하고, 정기적으로 비우며 청소한다.
공기 정화기 또는 여과 시스템의 사용을 고려한다.

7) 고객 공간의 위생
고객이 사용하는 치료대나 의자는 매 고객 후에 깨끗이 소독한다.
주기적으로 청소하고, 손 소독제를 고객이 쉽게 사용할 수 있도록 배치한다.

8) 감염 관리 및 예방
감염을 예방하기 위한 정책과 절차를 마련하고, 모든 직원이 이를 숙지하고 준수해야 한다.
직원이나 고객 중 감염 의심 증상이 있는 경우 즉시 대응 조치를 취해야 한다.
"""
```
